# Whimsy

To determine the "best" neighborhood, we created this sub-metric, "Whimsy", which is based on how entertaining a neighborhood is through how many parks it  has. In conclusion, the more playgrounds a neighborhood has, the higher its "Whimsy" score which is out of 20. In the data set below, the neighborhoods are extremely specific and hard to effectively categorize and analyze, therefore, we chose to analyze the maintenance responsibility column instead as each maintenance group covers a broader area.

```python
import pandas as pd

wh = pd.read_csv("playgrounds.csv")
wh.head(20)
```

the code below looks through the maintenance responsibility column and takes note of all unique variables in that column and the amount of times it appears in said column.

```python
sorted_counts = wh['maintenance_responsibility'].value_counts().sort_values(ascending=False)
print(sorted_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("playgrounds.csv")

# Count each instance of each responsibility in the 'Maintenance Responsibility' column
neighborhood_counts = df['Maintenance Responsibility'].value_counts()

# Bar graph
neighborhood_counts.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Maintenance Responsibility')
plt.ylabel('Count of Responsibilities')  # Corrected to a string label for the y-axis
plt.title('Count of Maintenance Responsibilities for Parks in Pittsburgh')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.show()
```
